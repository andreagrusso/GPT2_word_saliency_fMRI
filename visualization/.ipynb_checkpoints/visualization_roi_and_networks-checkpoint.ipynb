```python
import numpy as np
import pandas as pd
import os, pickle, csv, glob, copy
import nibabel as nb
from nilearn.plotting import  plot_surf_stat_map
from nilearn import datasets, surface
import matplotlib.pyplot as plt
import matplotlib
import glob 
```

```python
#change both the time window and the max t stat accordingly
#wdir = 'C:/Users/andre/Desktop/Salerno/projects/gpt/output'
wdir = '/mnt/c/Users/andre/Desktop/Salerno/projects/gpt/output'
#excel_file = os.path.join(wdir,'surp_pivot.xlsx')
excel_file = os.path.join(wdir,'importance_pivot.xlsx')
```

```python
table = pd.read_excel(excel_file, sheet_name = 'Intersection')
table
```

```python
atlas = nb.load(os.path.join('/mnt/c/Users/andre/Desktop/Salerno/projects/gpt/data/schaefer_2018/Schaefer2018_1000Parcels_7Networks_order_FSLMNI152_2mm.nii.gz'))
network_names = pd.read_excel(os.path.join('/mnt/c/Users/andre/Desktop/Salerno/projects/gpt/data/schaefer_2018/Network_names.xlsx'))
atlas4plot = datasets.fetch_atlas_schaefer_2018(n_rois=1000, yeo_networks=7, resolution_mm=2)

custom_cmap = matplotlib.cm.get_cmap('tab10',8)

measure = 'importance'#'surprisal'
```

Assign Network name to Nr of time window

```python
tmp_vec = np.zeros((6,7))


atlas_data = atlas.get_fdata()
from nilearn.image import new_img_like


#loop on the number of time windows
for i in range(6):
    
    atlas_mat = np.zeros_like(atlas_data)
    

    #loop over the table data
    for j in range(table.shape[0]):
        
        id_area = table.iloc[j]['ID']
        if table.iloc[j]['Nr of time window']==i+1: #if we are looking at the correct data
            network_id = network_names.loc[network_names['ID']==id_area]['ID Network'].values[0] #get the indices of the network
        
            atlas_mat[atlas_data == table.iloc[j]['ID']] = network_id*10
            tmp_vec[i,network_id-1]=tmp_vec[i,network_id-1]+1

    new_nii = new_img_like(atlas, atlas_mat) 

    new_nii.to_filename(os.path.join(wdir,'NrOfWindow_X_Network',measure,str(i+1)+'window_bonf05.nii.gz'))
    
print(tmp_vec)    
network_x_nr_of_win = pd.DataFrame(data=tmp_vec,columns=['Visual','SomatoMotor','Dorsal Attention', 'Salience/Ventral Attention',
                                          'Limbic', 'Control','Default'], index = np.arange(6)+1)
network_x_nr_of_win.to_excel(os.path.join(wdir,'NrOfWindow_X_Network','NrOfWindow_X_Network.xlsx'))

ax = network_x_nr_of_win.plot.bar(rot=0,figsize = (10,7), fontsize=30, colormap = custom_cmap)
ax.set_xlabel('Number of windows',size=30)
ax.set_ylabel('Number of ROIs',size=30)


plt.tight_layout()
plt.savefig(os.path.join(wdir,'NrOfWindow_X_Network',measure,'NrOfWindow_X_Network.png'),facecolor='white', dpi=300,transparent=False)
```

Assign Network name to each time window

```python
tmp_vec = np.zeros((6,7))


atlas_data = atlas.get_fdata()
from nilearn.image import new_img_like

time_windows = np.arange(15,105,15)
print(time_windows)

#loop on the number of time windows
for idx,i in enumerate(time_windows):
    
    atlas_mat = np.zeros_like(atlas_data)
    

    #loop over the table data
    for j in range(table.shape[0]):
        
        id_area = table.iloc[j]['ID']
        if  not np.isnan(table.iloc[j][i]): #if we are looking at the correct data
            network_id = network_names.loc[network_names['ID']==id_area]['ID Network'].values[0] #get the indices of the network
        
            atlas_mat[atlas_data == table.iloc[j]['ID']] = network_id*10
            tmp_vec[idx,network_id-1]=tmp_vec[idx,network_id-1]+1

            
    new_nii = new_img_like(atlas, atlas_mat) 

    new_nii.to_filename(os.path.join(wdir,'Window_X_Networks',measure,str(i)+'s_window_bonf05.nii.gz'))
    
    
print(tmp_vec)
network_x_nr_of_win = pd.DataFrame(data=tmp_vec,columns=['Visual','SomatoMotor','Dorsal Attention', 'Salience/Ventral Attention',
                                          'Limbic', 'Control','Default'],index=time_windows)    
network_x_nr_of_win.to_excel(os.path.join(wdir,'Window_X_Networks',measure,'Window_X_Networks.xlsx'))

ax = network_x_nr_of_win.plot.bar(rot=0, figsize = (10,7), fontsize=30, colormap = custom_cmap)
ax.set_xlabel('Seconds', size=30)
ax.set_ylabel('Number of ROIs',size=30)
plt.tight_layout()
plt.savefig(os.path.join(wdir,'Window_X_Networks', measure,'Window_X_Networks.png'),facecolor='white', dpi=300,transparent=False)
```

Nr of windows for each network

```python
from nilearn.plotting import plot_stat_map, plot_roi
from matplotlib.colors import ListedColormap

#custom_cmap =  ListedColormap(["blue", "orange","green", "red", "violet","brown","pink"])

nii_files = glob.glob(os.path.join(wdir,'NrOfWindow_X_Network',measure,'*.nii.gz'))
sorted(nii_files)


for i,curr_nii in enumerate(nii_files):
    
    curr_nii_data =  nb.load(os.path.join(curr_nii))
    
    display = plot_roi(curr_nii_data,
                            display_mode='z', 
                            cut_coords = 10, 
                            annotate=True, 
                            black_bg=False,
                            colorbar=True,
                            vmin = 10,
                            resampling_interpolation = 'nearest',
                            cmap = custom_cmap)#plt.cm.get_cmap('jet', 7))
                           
    display.add_contours(curr_nii_data, filled=False, alpha=.7,colors=['black'])
    display.annotate(size=14)
    display.savefig(os.path.join(wdir,'NrOfWindow_X_Network',
                                 measure,
                                 str(i+1)+'_time_window_networks_bonf05.svg'),dpi=600)    
```
